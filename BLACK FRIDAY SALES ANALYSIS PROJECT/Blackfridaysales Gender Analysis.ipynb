```python
df.groupby('Gender').size()
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]}, 
                    index = ['Male', 'Female'])

data.plot.pie(y = 'Ratio', figsize = (6,6), autopct = "%.1f") 
```

```python
data = df['Gender'].value_counts() # SIMPLER VERSION OF CODE USING ".value_counts()".

data.plot.pie(y = 'Ratio', figsize = (6,6), autopct = "%.1f", labels = ['Male', 'Female'],colors=['skyblue','pink'])
```

```python
df.groupby('Gender').size().plot(kind = 'pie', 
                                 autopct = "%.1f",
                                 title = 'Gender Ratio',
                                 figsize = (6,6),colors=['yellow','green'])
```
