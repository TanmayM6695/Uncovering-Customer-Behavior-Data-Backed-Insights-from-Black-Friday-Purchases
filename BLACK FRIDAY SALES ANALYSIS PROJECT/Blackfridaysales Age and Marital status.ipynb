```python
import os
import pandas as pd
df = pd.read_csv(r"C:\Users\HP\Downloads\PROJECTS\DATASETS\PYTHON\BlackFriday.csv")
del df['Product_Category_2']
del df ['Product_Category_3']
```

```python
df.groupby('Age').size().plot(kind='bar', figsize=(12,6), title=('Purchase Distribution by Age'))
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst , columns = ['Age','Products']) # LOOPS AND FILTERS THROUGH DATA AS PER LOOP CONDITION
```

```python
data.plot.bar(x='Age', figsize=(9,6))
```
